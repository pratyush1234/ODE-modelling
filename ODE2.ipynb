```python
import numpy as np
import matplotlib.pyplot as plt
dt = 0.1
U0 = 2
T = 4
N = int(T/dt)
t = np.zeros(N+1)
u = np.zeros(N+1)
t[0] = 0
u[0] = U0
for n in range(N):
  t[n+1] = t[n] + dt
  u[n+1] = (1+dt)*u[n]
  plt.plot(t,u)
  plt.show()




```
